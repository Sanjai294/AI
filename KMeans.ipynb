```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
num_clusters = int(input("Enter the number of clusters: "))
num_points = int(input("Enter the number of data points: "))
```

```python
points = []
for i in range(num_points):
    x = float(input(f"Enter x-coordinate for point {i + 1}: "))
    y = float(input(f"Enter y-coordinate for point {i + 1}: "))
    points.append([x, y])
```

```python
X = np.array(points)
```

```python
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
print("\nCluster Labels:")
for i, label in enumerate(labels):
    print(f"Point {i + 1} -> Cluster {label}")
```

```python
print("\nCluster Centers:")
for idx, center in enumerate(centers):
    print(f"Cluster {idx}: Center at {center}")
```
